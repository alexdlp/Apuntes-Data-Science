# 2 Anantomía de una figura en Matplotlib
Matplotlib genera sus datos en *Figures* (ventanas o lienzos en blanco), cada una de las cuales puede contener uno o más *Axes*

![title](https://matplotlib.org/3.1.1/_images/sphx_glr_anatomy_001.png)

# 2.1 Figura (*figure*)
Es todo el conjunto (ejes, títulos, leyendas, lienzo...). Una figura puede contener cualquier número de ejes, pero normalmente tendrá al menos uno.

# 2.2 Ejes (*axes*)
Es el gráfico propiamente dicho, la región de la imagen con datos. Una figura determinada puede contener muchos ejes, epro un eje determinado solo puede estar en una figura.
Los ejes a su vez contienen dos (o tres en el caso de 3D) objetos *Axis* (en Matplotlib, *Axes* y *Axis* hacen referencia a objetos distintos) que se encargan de los límites de datos (los límites de datos también se pueden controlar mediante los métodos *axes*. *Axes.set_xlim()* y *Axes.set_ylim())*.
Cada eje tiene un título (establecido mediante *set_title()*), una etiqueta *x* (establecida mediante *set_xlabel()*) y una etiqueta y (*set_ylabel()*).

# 2.3 Ejes (*axis*)
Son objetos parecidos a una recta numérica que se encargan de establecer los límites del gráfico y generar los ticks y los ticklabels (cadenas de texto que hacen referencia a cada tick)

# 2.4 Objetos *artist*
Básicamente todo lo que se puede ver en la figura en un *artist* (incluso los objetos de tipo Figure, Axes y Axis). Cuando se renderiza la figura, todos los *artists* se dibujan en el lienzo. La mayoría de los *artists* estçan atados a ejes; un artist no puede ser compartido por varios ejes, ni moverse de uno a otro.

# 3 La interfaz orientada a objetos y la intefaz pyplot

Hay esencialmente 2 formas de usar Matplotlib:
+ Creando explícitamente figuras y ejes, y llamando a métodos sobre ellos 
+ Confiando en pyplot para crear y administrar automáticamente las figuras y los ejes, y utilizando las funciones de pyplot para trazar

Cualquiera de ellos es perfectamente válido, aunque es preferible elegir uno de ellos y ceñirse a el, en lugar de mezclarlos en un mismo script o cuaderno de Jupyter. En general, es recomendable restrignir pyplot al trazado interactivo (por ejemplo, en un cuaderno Jupyter) y preferir la OOP para el trazado no interactivo (en fuciones y scripts que están destinados a ser reutilizados como parte de un proyecto más grande).

# 3.1 Estilo OOP

```python
x = np.linspace(0, 2, 100)
```

```python
# una figura vacía sin ejes
fig = plt.figure()
```

```python
# una figura con un solo eje
fig, ax = plt.subplots()
ax.plot(x, x, label = "lineal")
ax.plot(x, x**2, label = "cuadrático")
ax.plot(x, x**3, label = "cúbico")
ax.set_xlabel("Etiqueta eje X")
ax.set_ylabel("Etiqueta eje Y")
ax.set_title("Gráfico de ejemplo")
ax.legend()
```

```python
# Una figura con una cuadrícula de 2x2 ejes (4 ejes)
fig, axes = plt.subplots(2,2)
```

# 3.2 Estilo pyplot

```python
plt.plot(x, x, label = "lineal")
plt.plot(x, x**2, label = "cuadrático")
plt.plot(x, x**3, label = "cúbico")
plt.xlabel("Etiqueta eje X")
plt.ylabel("Etiqueta eje Y")
plt.title("Gráfico de ejemplo")
plt.legend()
```

# 4 Plantilla de función
Es habitual utilizar los mismos gráficos una y otra vez, pero con diferentes conjuntos de datos, lo que lleva a la necesidad de escribir funciones especializadas 

```python
def plot_funcion(ax, data1, data2, param_dict):
    """Una función de ayuda para hacer un gráfico con matplotlib.pyplot

    Args:
        ax (Axes): Los ejes sobre los cuales dibujar
        data1 (array): un primer vector de datos (x)
        data2 (array): Un segundo vector de datos (y)
        param_dict (dict): Diccionario de parámetros para pasarle a ax.plot
    Returns:
        (list) : Lista de objetos de tipo "artists" añadidos
    """

    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
fig, ax = plt.subplots(1,1)
data1 = [1,2,3,4]
data2 = [2,3,4,5]
par_dict = {"marker": "o", "color":"green", "markersize":12}

plot_funcion(ax = ax, data1 = data1, data2 = data2, param_dict = par_dict)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
plot_funcion(ax = ax1, data1 = data1, data2 = data2, param_dict = {"marker": "o", "color":"red", "markersize":6})
plot_funcion(ax = ax2, data1 = data1, data2 = data2, param_dict = {"marker": "o", "color":"green", "markersize":12})
```
